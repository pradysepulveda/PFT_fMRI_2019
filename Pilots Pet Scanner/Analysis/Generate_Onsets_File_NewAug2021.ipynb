```python
import numpy as np
import pandas as pd
from pandas.core.frame import DataFrame as DF
# Useful to quickly extract csv files
import glob
import os
```

```python
def z_score_part(data_all, part_def,z_score_var):
    z_matrix=[]
    z_matrix_aux=[]

    for i in (data_all[part_def].unique()):
        Choicedata = data_all.loc[data_all[part_def] == i]    
    
        pX_A= pd.to_numeric(Choicedata[z_score_var]) 
        pX_zA= (pX_A - np.mean(pX_A))/np.std(pX_A)

        z_matrix_aux= pX_zA.values
    
        for  j in range(len(z_matrix_aux)):    
            z_matrix.append(z_matrix_aux[j])
    return z_matrix
```

```python
def z_score_1(data_all,z_score_var):
    z_matrix=[]
    z_matrix_aux=[]
    
    pX_A= pd.to_numeric(data_all[z_score_var]) 
    pX_zA= (pX_A - np.mean(pX_A))/np.std(pX_A)

    z_matrix= pX_zA.values
    
    return z_matrix
```

```python
partic_num=len(rateNameFrames) #number of participants

rateFrameRawAll = pd.DataFrame( )
choiceFrameRawAll = pd.DataFrame( )
expTimeFrameRawAll = pd.DataFrame( )
audTimeFrameRawAll = pd.DataFrame( )

for j in range (partic_num):
    
    
    rateFrames = pd.read_csv(rateNameFrames[j])
    choiceFrames = pd.read_csv(choiceNameFrames[j]) 
    expSummFrames = pd.read_csv(expSummaryFrames[j],names=['timeLabel','time'], header=None)
    audioFrames = pd.read_csv(audioTimeFrames[j],names=['TestAudioTime'], header=None)
   # print (rateFrames.rating)
    
    if 'ratL' in choiceFrames.columns:
        if 'ratR' in choiceFrames.columns:
            choiceFrames = choiceFrames.drop(['ratL', 'ratR'], axis=1)
   
    partNamesRate = [int(os.path.basename(choiceNameFrames[j])[1:3])]*len(rateFrames)
    partNamesChoice = [int(os.path.basename(rateNameFrames[j])[1:3])]*len(choiceFrames)
    partNamesExp = [int(os.path.basename(rateNameFrames[j])[1:3])]*len(expSummFrames)
    partNamesAudio = [int(os.path.basename(rateNameFrames[j])[1:3])]*len(audioFrames)

    
    sessionRate = [int(os.path.basename(choiceNameFrames[j])[-5]) +1 ]*len(rateFrames)
    sessionChoice = [int(os.path.basename(rateNameFrames[j])[-5]) +1 ]*len(choiceFrames)
    sessionExpTime = [int(os.path.basename(expSummaryFrames[j])[-5]) +1 ]*len(expSummFrames)
    sessionAudTime = [int(os.path.basename(audioTimeFrames[j])[-5]) +1 ]*len(audioFrames)
    
    # get names and session number
    rateFrames['part'] = partNamesRate 
    choiceFrames['part'] = partNamesChoice 
    expSummFrames['part'] = partNamesExp 
    audioFrames['part'] = partNamesAudio 

    rateFrames['session'] = sessionRate 
    choiceFrames['session'] = sessionChoice 
    expSummFrames['session'] = sessionExpTime 
    audioFrames['session'] = sessionAudTime 
    
   # # to fix rating in wrong participant (only participant 1)
   # if np.unique(partNamesRate) == 1 and np.unique(sessionRate)[0]%2 == 1: # participant 1 is the conflictive one.
   #     ratingFixed = 5 - rateFrames.rating.values
   #     rateFrames['rating'] = [-1 if x==6 else x for x in ratingFixed]
   #     print('into')

    # change trial time according to trigger    
    triggerTime =  expSummFrames.loc[(expSummFrames['timeLabel'] == 'timeTrigger')].time.values[0]
    
    # trialOnset
    choiceFrames['trialOnset'] =  choiceFrames.trialOnset - triggerTime
    rateFrames['trialOnset'] =  rateFrames.trialOnset - triggerTime
    audioFrames['TestAudioTime'] = audioFrames.TestAudioTime - triggerTime

    # update timings for rate data
    rateFrames['trialImageOnst'] =  rateFrames.trialOnset + rateFrames.trialImageOnst 
    rateFrames['trialRateOnst'] =  rateFrames.trialOnset*(rateFrames.trialRateOnst >0)  + rateFrames.trialRateOnst # add only if it is a rate trial
    rateFrames['trialRatEnd'] =  rateFrames.trialOnset*(rateFrames.trialRatEnd >0) + rateFrames.trialRatEnd # add only if it is a rate trial
    rateFrames['trialEnd'] =  rateFrames.trialOnset + rateFrames.trialEnd

    # update timings for choice data
    choiceFrames['trialOnsetTim'] =  choiceFrames.trialOnset + choiceFrames.trialOnsetTim
    choiceFrames['trialSnd1Tim'] =  choiceFrames.trialOnset + choiceFrames.trialSnd1Tim
    choiceFrames['trialSnd2Tim'] =  choiceFrames.trialOnset + choiceFrames.trialSnd2Tim
    choiceFrames['trialThinkTime'] =  choiceFrames.trialOnset + choiceFrames.trialThinkTime
    choiceFrames['trialRespTime'] =  choiceFrames.trialOnset + choiceFrames.trialRespTime
    choiceFrames['trialEnd'] =  choiceFrames.trialOnset + choiceFrames.trialEnd
    choiceFrames['choiceTime'] =  choiceFrames.trialOnset + choiceFrames.choiceTime
    
    ## join the dataframes
    choiceFrameRawAll = pd.concat([choiceFrameRawAll,choiceFrames],ignore_index=True)
    rateFrameRawAll = pd.concat([rateFrameRawAll,rateFrames],ignore_index=True)
    expTimeFrameRawAll = pd.concat([expTimeFrameRawAll,expSummFrames],ignore_index=True)
    audTimeFrameRawAll = pd.concat([audTimeFrameRawAll,audioFrames],ignore_index=True)
    
```

## add value data to the choice dataframe 

```python
choiceFrameAll = pd.DataFrame()
idFrameAll = pd.DataFrame( )
rateFrameAll = pd.DataFrame( )
```

```python
for j in list (choiceFrameRawAll.part.unique()):
    
    idNames = list(rateFrameRawAll.picId.unique())
    IdRatings1 = [None]*len(idNames)
    IdRatings2 = [None]*len(idNames)
    IdRatingsVar1 = [None]*len(idNames)
    IdRatingsVar2 = [None]*len(idNames)

    #extract average rating for like, dislike
    ###########################################
    for i in range(len(idNames)):
        IdRatings1[i] = rateFrameRawAll.loc[(rateFrameRawAll['picId'] == idNames[i]) & (rateFrameRawAll['frame'] == 1) & (rateFrameRawAll['part'] == j) & (rateFrameRawAll['rating'] >= 0)].rating.mean()
        IdRatings2[i] = rateFrameRawAll.loc[(rateFrameRawAll['picId'] == idNames[i]) & (rateFrameRawAll['frame'] == 2) & (rateFrameRawAll['part'] == j) & (rateFrameRawAll['rating'] >= 0)].rating.mean()

        IdRatingsVar1[i] = rateFrameRawAll.loc[(rateFrameRawAll['picId'] == idNames[i]) & (rateFrameRawAll['frame'] == 1) & (rateFrameRawAll['part'] == j) & (rateFrameRawAll['rating'] >= 0)].rating.std()
        IdRatingsVar2[i] = rateFrameRawAll.loc[(rateFrameRawAll['picId'] == idNames[i]) & (rateFrameRawAll['frame'] == 2) & (rateFrameRawAll['part'] == j) & (rateFrameRawAll['rating'] >= 0)].rating.std()
    
    rateValues = pd.DataFrame( {'nameId': idNames,'rat1': IdRatings1,'rat2': IdRatings2,'var1': IdRatingsVar1,'var2': IdRatingsVar2,})
    rateValues['part'] = [j]*len(rateValues)    
    
    # save information about animal and ratings
    ###########################################
    
    idFrameAll = pd.concat([idFrameAll,rateValues],ignore_index=True)
    
    #  add rating info into the rate dataframe
    ###########################################
    rateFrames = rateFrameRawAll.loc[ (rateFrameRawAll['part'] == j)]

    RateItmVal1 =  [None]*len(rateFrames)
    RateItmVal2 =  [None]*len(rateFrames)
    RateItmVar1 =  [None]*len(rateFrames)
    RateItmVar2 =  [None]*len(rateFrames)
    RateExemplarVal =  [None]*len(rateFrames)
     
    for i in range(len(rateFrames)):
        
        rateItm = rateFrames.iloc[i].picId
        rateItmNum = rateFrames.iloc[i].picNum
        
        if any((rateValues['nameId'] == rateItm)) :
        
            RateItmVal1[i] = rateValues.loc[(rateValues['nameId'] == rateItm)].rat1.values[0]
            RateItmVal2[i] = rateValues.loc[(rateValues['nameId'] == rateItm)].rat2.values[0]
            
            RateItmVar1[i] = rateValues.loc[(rateValues['nameId'] == rateItm)].var1.values[0]
            RateItmVar2[i] = rateValues.loc[(rateValues['nameId'] == rateItm)].var2.values[0]
            
            RateExemplarVal[i] = rateFinalRawAll.loc[(rateFinalRawAll['picId'] == rateItm) & (rateFinalRawAll['picNum'] == rateItmNum)  & (rateFinalRawAll['part'] == j)].rating.values[0]

    rateFrames['Val1'] = RateItmVal1
    rateFrames['Val2'] = RateItmVal2
    rateFrames['Var1'] = RateItmVar1
    rateFrames['Var2'] = RateItmVar2
    rateFrames['ExVal'] = RateExemplarVal
    
    # extra variables  -- rate
    ###########################################
    # z-score
    rateFrames['zVal1'] = z_score_1(rateFrames,'Val1')
    rateFrames['zVal2'] = z_score_1(rateFrames,'Val2')
    rateFrames['zVar1'] = z_score_1(rateFrames,'Var1')
    rateFrames['zVar2'] = z_score_1(rateFrames,'Var2')
    rateFrames['zExVal'] = z_score_1(rateFrames,'ExVal')

    # add rating info into the choice dataframe
    ###########################################
    choiceFrames = choiceFrameRawAll.loc[ (choiceFrameRawAll['part'] == j)]
    
    
    # separate ratings and rating-variability for each frame
    ChoiceLRat1 = [None]*len(choiceFrames)
    ChoiceRRat1 = [None]*len(choiceFrames)
    ChoiceLRat2 = [None]*len(choiceFrames)
    ChoiceRRat2 = [None]*len(choiceFrames)

    ChoiceLVar1 = [None]*len(choiceFrames)
    ChoiceRVar1 = [None]*len(choiceFrames)
    ChoiceLVar2 = [None]*len(choiceFrames)
    ChoiceRVar2 = [None]*len(choiceFrames)
    
    for i in range(len(choiceFrames)):
        
        choiceL = choiceFrames.iloc[i].leftId
        choiceR = choiceFrames.iloc[i].rightId
        
        if any((rateValues['nameId'] == choiceL)) and any((rateValues['nameId'] == choiceR)):
        
            ChoiceLRat1[i] = rateValues.loc[(rateValues['nameId'] == choiceL)].rat1.values[0]
            ChoiceRRat1[i] = rateValues.loc[(rateValues['nameId'] == choiceR)].rat1.values[0]
            ChoiceLRat2[i] = rateValues.loc[(rateValues['nameId'] == choiceL)].rat2.values[0]
            ChoiceRRat2[i] = rateValues.loc[(rateValues['nameId'] == choiceR)].rat2.values[0]
            
            ChoiceLVar1[i] = rateValues.loc[(rateValues['nameId'] == choiceL)].var1.values[0]
            ChoiceRVar1[i] = rateValues.loc[(rateValues['nameId'] == choiceR)].var1.values[0]
            ChoiceLVar2[i] = rateValues.loc[(rateValues['nameId'] == choiceL)].var2.values[0]
            ChoiceRVar2[i] = rateValues.loc[(rateValues['nameId'] == choiceR)].var2.values[0]

    choiceFrames['LRat1'] = ChoiceLRat1
    choiceFrames['RRat1'] = ChoiceRRat1
    choiceFrames['LRat2'] = ChoiceLRat2
    choiceFrames['RRat2'] = ChoiceRRat2 
    
    choiceFrames['LVar1'] = ChoiceLVar1
    choiceFrames['RVar1'] = ChoiceRVar1
    choiceFrames['LVar2'] = ChoiceLVar2
    choiceFrames['RVar2'] = ChoiceRVar2
    
    # add extra variables  -- choice
    ###########################################
    
    choiceFrames["TotVal1"] = choiceFrames['RRat1']+ choiceFrames['LRat1']
    choiceFrames["TotVal2"] = choiceFrames['RRat2']+ choiceFrames['LRat2']


    choiceFrames["DVal1"] = choiceFrames['RRat1'] - choiceFrames['LRat1']
    choiceFrames["absDVal1"] = np.abs(choiceFrames["DVal1"] )

    choiceFrames["DVar1"] = choiceFrames['RVar1'] - choiceFrames['LVar1']
    choiceFrames["absDVar1"] = np.abs(choiceFrames["DVar1"] )

    choiceFrames["DVal2"] = choiceFrames['RRat2'] - choiceFrames['LRat2']
    choiceFrames["absDVal2"] = np.abs(choiceFrames["DVal2"] )

    choiceFrames["DVar2"] = choiceFrames['RVar2'] - choiceFrames['LVar2']
    choiceFrames["absDVar2"] = np.abs(choiceFrames["DVar2"] )
    
    
    choiceFrames["ChosenVal1"] = choiceFrames['RRat1']*(choiceFrames['choice']+1)/2+ choiceFrames['LRat1']*(1-choiceFrames['choice'])/2
    choiceFrames["UnchosenVal1"] = choiceFrames['LRat1']*(choiceFrames['choice']+1)/2+ choiceFrames['RRat1']*(1-choiceFrames['choice'])/2
    choiceFrames["DChoUnchosen1"] = choiceFrames['ChosenVal1'] - choiceFrames['UnchosenVal1']

    
    choiceFrames["ChosenVal2"] = choiceFrames['RRat2']*(choiceFrames['choice']+1)/2+ choiceFrames['LRat2']*(1-choiceFrames['choice'])/2
    choiceFrames["UnchosenVal2"] = choiceFrames['LRat2']*(choiceFrames['choice']+1)/2+ choiceFrames['RRat2']*(1-choiceFrames['choice'])/2
    choiceFrames["DChoUnchosen2"] = choiceFrames['ChosenVal2'] - choiceFrames['UnchosenVal2']

    
    # z-score the variables
    choiceFrames['zLRat1'] =     z_score_1(choiceFrames,'LRat1')
    choiceFrames['zRRat1'] =     z_score_1(choiceFrames,'RRat1')
    choiceFrames['zLRat2'] =     z_score_1(choiceFrames,'LRat2')
    choiceFrames['zRRat2'] =     z_score_1(choiceFrames,'RRat2')
 
    
    choiceFrames['zLVar1'] =     z_score_1(choiceFrames,'LVar1')
    choiceFrames['zRVar1'] =     z_score_1(choiceFrames,'RVar1')
    choiceFrames['zLVar2'] =     z_score_1(choiceFrames,'LVar2')
    choiceFrames['zRVar2'] =     z_score_1(choiceFrames,'RVar2')
    
    choiceFrames["zDVal1"] = z_score_1(choiceFrames,"DVal1")
    choiceFrames["zabsDVal1"] = z_score_1(choiceFrames,"absDVal1")
    
    choiceFrames["zDVal2"] = z_score_1(choiceFrames,"DVal2")
    choiceFrames["zabsDVal2"] = z_score_1(choiceFrames,"absDVal2")
    
    choiceFrames["zDVar1"] = z_score_1(choiceFrames,"DVar1")
    choiceFrames["zabsDVar1"] = z_score_1(choiceFrames,"absDVar1")
    
    choiceFrames["zDVar2"] = z_score_1(choiceFrames,"DVar2")
    choiceFrames["zabsDVar2"] = z_score_1(choiceFrames,"absDVar2")
    
    choiceFrames["zTotVal1"] = z_score_1(choiceFrames,"TotVal1")
    choiceFrames["zTotVal2"] = z_score_1(choiceFrames,"TotVal2")
    choiceFrames["zChoiceTime"] = z_score_1(choiceFrames,"choiceTime")
    
    choiceFrames["zChosenVal1"] = z_score_1(choiceFrames,"ChosenVal1")
    choiceFrames["zUnchosenVal1"] = z_score_1(choiceFrames,"UnchosenVal1")
    choiceFrames["zChosenVal2"] = z_score_1(choiceFrames,"ChosenVal2")
    choiceFrames["zUnchosenVal2"] = z_score_1(choiceFrames,"UnchosenVal2")

    choiceFrames["zDChoUnchosen1"] = z_score_1(choiceFrames,"DChoUnchosen1")
    choiceFrames["zDChoUnchosen2"] = z_score_1(choiceFrames,"DChoUnchosen2")

    
    #choiceFrames['Part'] = [j]*len(choiceFrames)    
    rateFrameAll = pd.concat([rateFrameAll,rateFrames],ignore_index=True)
    choiceFrameAll = pd.concat([choiceFrameAll,choiceFrames],ignore_index=True)
    
```
